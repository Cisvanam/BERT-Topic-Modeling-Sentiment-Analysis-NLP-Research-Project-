```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histogram for numerical columns
df.hist(figsize=(12, 10), bins=20)
plt.tight_layout()
plt.show()

# Boxplots for numerical columns
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot for {column}')
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot top 20 value counts for 'city' and 'state' columns
for column in ['city', 'state']:
    plt.figure(figsize=(6, 6))
    df[column].value_counts().head(20).plot(kind='bar', color='lightblue')
    plt.title(f'Top 20 Distribution of {column}')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

# Plot top 20 value counts for 'categories' column
top_categories = df['categories'].str.split(',').explode().value_counts().head(20)
top_categories.plot(kind='bar', figsize=(12, 6), color='lightgreen')
plt.title('Top 20 Categories Distribution')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
